```python
import nbformat

filename = "/content/drive/MyDrive/Colab Notebooks/research_porfolio/result/Graph_viz4.ipynb"  # ここを自分のファイル名に変更
nb = nbformat.read(filename, as_version=nbformat.NO_CONVERT)

widgets_key = "application/vnd.jupyter.widget-state+json"
widgets = nb.metadata.get("widgets", {})

if widgets_key in widgets:
    if "state" not in widgets[widgets_key]:
        widgets[widgets_key]["state"] = {}
        print("✅ Added missing 'state' key.")
else:
    # widgetセクションごと不完全な場合
    nb.metadata["widgets"] = {
        widgets_key: {
            "state": {},
            "version_major": 2,
            "version_minor": 0
        }
    }
    print("✅ Reconstructed 'widgets' metadata section.")

nbformat.write(nb, filename)
```
